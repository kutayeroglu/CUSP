```python
# --------------------------- CONFIG ---------------------------------
ARROW_DIR = "/content/drive/MyDrive/datasets/FFHQaging"          # 16 shards
CSV_PATH  = "/content/drive/MyDrive/datasets/FFHQaging_LS/FFHQaging.csv"
IMG_DIR   = "/content/drive/MyDrive/datasets/FFHQaging_LS/images"

# --------------------------- SET‑UP ---------------------------------
from datasets import load_from_disk
import pandas as pd, os
from PIL import Image
from tqdm.auto import tqdm

os.makedirs(IMG_DIR, exist_ok=True)

# labels – only to grab the expected file names
df = pd.read_csv(CSV_PATH)
fname_series = df["fname"]                    # length 70 k

# images
ds = load_from_disk(ARROW_DIR)
assert len(ds) == len(fname_series), "Dataset/CSV length mismatch!"

# --------------------------- EXPORT ---------------------------------
for idx, example in tqdm(enumerate(ds), total=len(ds), desc="Saving PNGs"):
    out_path = os.path.join(IMG_DIR, fname_series.iloc[idx])

    # skip if the PNG already exists (handy if the loop was interrupted)
    if os.path.exists(out_path):
        continue

    # `example["image"]` is a PIL Image
    img: Image.Image = example["image"].convert("RGB")  # ensure 3‑channel
    img.save(out_path, format="PNG")

print("All images written to", IMG_DIR)
```
